```python
#SORTING MATRIX PRESERVING INDEX AND SIMILARITY SCORE
def Sort_Tuple(tup):   
    tup.sort(key = lambda x: x[1],reverse=True)  
    return tup
```

```python
def difference(A):
    diff=[]
    for i in range(0,len(A)):
        temp=[]
        for j in range(0,len(A[i])):
             temp.append((j,float(A[i][j])))
        diff.append(temp)
    for i in range(0,len(diff)):
       Sort_Tuple(diff[i])  
    return diff
```

```python
#WEIGHTED AVERAGE 
w_d=[]
eta=0.7
sum_d=[]
n_dd_dnn=[]
for i in range (0,size_d):
    print(i)
    w=0.0
    s=0.0
    for j in range (0,len(pnei_d_num[i])):
        w+=(pow(eta,j)*pnei_d_num[i][j])
        s+=pnei_d_num[i][j]
    w_d.append(w)
    sum_d.append(s)
for i in range(0,size_d):
    temp=0.0
    for j in range(0,len(pnei_d[i])):
        for k in range(0,size_t): 
            temp+=(w_d[i] * float(n_dt_2[k][pnei_d[i][j]]))
    if(sum_d[i]>0):
        temp=temp/sum_d[i]
    n_dd_dnn.append(temp)
```

```python
# taking 100 iterations for convergence 
l_d=0.5
l_t=0.5
l_l=0.5

for i in range (0,100):
    temp_1=np.multiply(n_dt_2[0:50,0:50],B) 
    temp_2=np.multiply(n_dd_3[0:50,0:50],A)            
    temp_2=l_d*temp_2
    temp_3=(np.transpose(B)).dot(B)
    x,y=np.shape(temp_3)
    temp_4=(np.identity(x,dtype=float))
    temp_4=l_l*temp_4
    temp_5=(np.transpose(n_dt_2[0:50,0:50])).dot(A)
    temp_6=n_tt_3[0:50,0:50].dot(B)
    temp_6=l_t*temp_6
    temp_7=(np.transpose(A)).dot(A)
    x1,y1=np.shape(temp_7)
    temp_8=(np.identity(x1,dtype=float))
    temp_8=l_l*temp_8
    A=(temp_1-temp_2).dot(np.linalg.pinv(temp_3+temp_4))
    B=(temp_5-temp_6).dot(np.linalg.pinv(temp_8+temp_7))
    #np.linalg.pinv(a)
```
